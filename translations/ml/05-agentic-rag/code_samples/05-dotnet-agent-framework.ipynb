# 🔍 Microsoft Foundry (.NET) ഉപയോഗിച്ചുള്ള എന്റർപ്രൈസ് RAG

## 📋 പഠന ലക്ഷ്യങ്ങൾ

ഈ നോട്ട്‌ബുക്ക് Microsoft Foundry ഉപയോഗിച്ച് .NET ൽ Microsoft Agent Framework ഉപയോഗിച്ച് എന്റർപ്രൈസ്-ഗ്രേഡ് Retrieval-Augmented Generation (RAG) സിസ്റ്റങ്ങൾ എങ്ങനെ നിർമ്മിക്കാമെന്ന് കാണിക്കുന്നു. നിങ്ങൾ പ്രൊഡക്ഷൻ-റെഡിയായി ഉപയോഗിക്കാൻ കഴിയുന്ന ഏജന്റുകൾ സൃഷ്ടിക്കാൻ പഠിക്കും — ഇവ ഡോക്യുമെന്റുകൾ തിരയുകയും എന്റർപ്രൈസ് സുരക്ഷയും സ്കേലബിലിറ്റിയും ഉൾക്കൊള്ളുന്ന കൃത്യവും സാഹചര്യബോധമുള്ള മറുപടികൾ നൽകുകയും ചെയ്യും.

**നിങ്ങൾ നിർമ്മിക്കാൻ പോകുന്ന എന്റർപ്രൈസ് RAG ശേഷികൾ:**
- 📚 **Document Intelligence**: Azure AI സേവനങ്ങൾ ഉപയോഗിച്ചുള്ള ഉയർന്ന നിലവാരമുള്ള ഡോക്യുമെന്റ് പ്രോസസ്സിംഗ്
- 🔍 **Semantic Search**: എന്റർപ്രൈസ് ഫീച്ചറുകളോടെ ഉയർന്ന പ്രകടന ശേഷിയുള്ള വെക്ടർ തിരയൽ
- 🛡️ **Security Integration**: റോൾ-അധിഷ്‌ഠിത ആക്‌സസ് ಮತ್ತು ഡാറ്റ സംരക്ഷണ മാതൃകകൾ
- 🏢 **Scalable Architecture**: നിരീക്ഷണങ്ങളോടുകൂടെ പ്രൊഡക്ഷൻ-റഡിയായ RAG സിസ്റ്റങ്ങൾ

## 🎯 എന്റർപ്രൈസ് RAG ആർക്കിടെക്ചർ

### പ്രധാന എന്റർപ്രൈസ് ഘടകങ്ങൾ
- **Microsoft Foundry**: സുരക്ഷയും പാലനവും ഉൾക്കൊള്ളുന്ന മാനേജുചെയ്ത എന്റർപ്രൈസ് AI പ്ലാറ്റ്ഫോം
- **Persistent Agents**: സംഭാഷണ ചരിത്രവും കോൺടെക്സ് മാനേജ്മെന്റും ഉള്ള സ്റ്റേറ്റ്‌ഫുൾ ഏജന്റുകൾ
- **Vector Store Management**: എന്റർപ്രൈസ്-നിരപ്പിലുള്ള ഡോക്യുമെന്റ് ഇൻഡെക്സ് ചെയ്യലും റിട്രീവലും
- **Identity Integration**: Azure AD പ്രാമാണീകരണവും റോൾ-അധിഷ്‌ഠിത ആക്‌സസ് കൺട്രോളും

### .NET എന്റർപ്രൈസ് ആനുകൂല്യങ്ങൾ
- **Type Safety**: RAG ഓപ്പറേഷനുകളുടെയും ഡാറ്റ സ്‌ട്രക്ചറുകളുടെയും കമ്പൈൽ-ടൈം പരിശോധനം
- **Async Performance**: ബ്ലോക്ക് ചെയ്യാത്ത ഡോക്യുമെന്റ് പ്രോസസിങ്ങും തിരയൽ ഓപ്പറേഷനുകളും
- **Memory Management**: വലിയ ഡോക്യുമെന്റ് ശേഖരങ്ങൾക്ക് കാര്യക്ഷമമായ റിസോഴ്‌സ് ഉപയോഗം
- **Integration Patterns**: ഡിപ്പെൻഡൻസി ഇൻജക്ഷൻ ഉപയോഗിച്ച് നേറ്റീവ് Azure സേവനങ്ങളുടെ സംയോജനം

## 🏗️ സാങ്കേതിക ആർക്കിടെക്ചർ

### എന്റർപ്രൈസ് RAG പൈപ്പ്ലൈൻ
```csharp
Document Upload → Security Validation → Vector Processing → Index Creation
                      ↓                    ↓                  ↓
User Query → Authentication → Semantic Search → Context Ranking → AI Response
```

### അടിസ്ഥാന .NET ഘടകങ്ങൾ
- **Azure.AI.Agents.Persistent**: സ്റ്റേറ്റ് നിലനിർത്തലോടെയുള്ള എന്റർപ്രൈസ് ഏജന്റ് മാനേജ്മെന്റ്
- **Azure.Identity**: സുരക്ഷിത Azure സേവന ആക്‌സസിനുള്ള ഇന്റഗ്രേറ്റഡ് പ്രാമാണീകരണം
- **Microsoft.Agents.AI.AzureAI**: Azure-ഓപ്റ്റിമൈസ്ഡ് ഏജന്റ് ഫ്രെയിംവർക്ക് നടപ്പാക്കൽ
- **System.Linq.Async**: ഉയർന്ന പ്രകടന ശേഷിയുള്ള അസിന്ത്രോണസ് LINQ പ്രവർത്തനങ്ങൾ

## 🔧 എന്റർപ്രൈസ് ഫീച്ചറുകൾ & ആനുകൂല്യങ്ങൾ

### സുരക്ഷയും പാലനവും
- **Azure AD Integration**: എന്റർപ്രൈസ് ഐഡന്റിറ്റി മാനേജ്മെന്റും പ്രാമാണീകരണവും
- **Role-Based Access**: ഡോക്യുമെന്റ് ആക്‌സസ് மற்றும் ഓപ്പറേഷനുകൾക്കുള്ള സൂക്ഷ്മ അവകാശ നിയന്ത്രണം
- **Data Protection**: സെൻസിറ്റീവ് ഡോക്യുമെന്റുകൾക്ക് റെസ്റ്റ്-ഇൽ ആൻഡ് ട്രാൻസിറ്റ് എൻക്രിപ്ഷൻ
- **Audit Logging**: പാലന ആവശ്യങ്ങൾക്ക് അനുയോജ്യമായ സമഗ്ര പ്രവര്‍ത്തന ട്രാക്കിംഗ്

### പ്രകടനവും സ്കേലബിലിറ്റിയും
- **Connection Pooling**: Azure സേവന കണക്ഷനുകളുടെ കാര്യക്ഷമ മാനേജ്മെന്റ്
- **Async Processing**: ഉയർന്ന ത്രൂപുട്ട് സാഹചര്യങ്ങൾക്ക് ബ്ലോക്ക് ചെയ്യാത്ത ഓപ്പറേഷനുകൾ
- **Caching Strategies**: പലപ്പോഴും ആക്‌സസ് ചെയ്യുന്ന ഡോക്യുമെന്റുകൾക്കുള്ള ബുദ്ധിപൂർണ്ണ കാഷിംഗ്
- **Load Balancing**: വലിയ സ്‌കെയിലിലുള്ള ഡിപ്ലോയ്മെന്റുകൾക്കുള്ള വിതരിച്ച പ്രോസസ്സിങ്

### മാനേജ്മെൻ്റ് & നിരീക്ഷണം
- **Health Checks**: RAG സിസ്റ്റം ഘടകങ്ങളിലേക്കുള്ള ഇൻബിൽറ്റ് നിരീക്ഷണം
- **Performance Metrics**: തിരയൽ ഗുണനിലവാരത്തെയും മറുപടി സമയങ്ങളെയും കുറിച്ചുള്ള വിശദമായ അനലിറ്റിക്സ്
- **Error Handling**: റിട്രൈ നയങ്ങളോടെ സമഗ്ര എക്സപ്ഷൻ മാനേജ്മെന്റ്

## 📊 എന്റർപ്രൈസ് RAG മാതൃകകൾ

### ഡോക്യുമെന്റ് മാനേജ്‌മെന്റ് മാതൃകകൾ
- **Bulk Upload**: വലിയ ഡോക്യുമെന്റ് ശേഖരങ്ങളുടെ കാര്യക്ഷമമായ പ്രോസസ്സിംഗ്
- **Incremental Updates**: റിയൽ-ടൈം ഡോക്യുമെന്റ് ചേർക്കലുകൾക്കും മാറ്റങ്ങൾക്കും
- **Version Control**: ഡോക്യുമെന്റ് വേർഷനിംഗ് மற்றும் മാറ്റം ട്രാക്കിംഗ്
- **Metadata Management**: സമൃദ്ധമായ ഡോക്യുമെന്റ് ഗുണഗതികൾക്കും ടാക്സോണമിക്കും

### തിരയലും റിറ്റ്‌റീവലും മാതൃകകൾ
- **Hybrid Search**: മികച്ച ഫലങ്ങൾക്ക് സെമാന്റിക് ഒപ്പം കീവേഡ് തിരയൽ സംയോജിപ്പിക്കൽ
- **Faceted Search**: ബഹുഡൈമെൻഷണൽ ഫിൽറ്ററിംഗ് һәм വർഗ്ഗീകരണം
- **Relevance Tuning**: ഡൊമെയ്ൻ-നിർദിഷ്ട ആവശ്യങ്ങൾക്ക് കസ്റ്റം സ്കോറിംഗ് അല്ഗോരിതങ്ങൾ
- **Result Ranking**: ബിസിനസ് ലജിക് സംയോജിപ്പിച്ച് ഉന്നത റാങ്കിംഗ്

### സുരക്ഷാ മാതൃകകൾ
- **Document-Level Security**: ഓരോ ഡോക്യുമെന്റിനുള്ള സൂക്ഷ്മ ആക്‌സസ് നിയന്ത്രണം
- **Data Classification**: സ്വയം പ്രവർത്തിക്കുന്ന സെൻസിറ്റിവിറ്റി ലേബലിംഗ് এবং സംരക്ഷണം
- **Audit Trails**: എല്ലാ RAG ഓപ്പറേഷനുകളുടെയും സമഗ്ര ലോഗിംഗ്
- **Privacy Protection**: PII കണ്ടെത്തലും റീഡാക്ഷൻ ശേഷികളും

- **Access Controls**: ഉപയോക്താക്കളുടെയും ഗ്രൂപ്പുകളുടെയുമായ അനുമതികൾക്കായി Azure AD സംയോജനം
- **Data Residency**: പാലനത്തിനായി ഭൂമിശാസ്ത്ര-അടിസ്ഥാന ഡാറ്റ സ്ഥലം നിയന്ത്രണങ്ങൾ
- **Backup & Recovery**: ഓട്ടോമേറ്റഡ് ബാക്കപ്പ് ഒപ്പം തകർച്ചശേഷ പുനരുദ്ധാരണ ശേഷികൾ

### മെമ്മറി മാനേജ്മെന്റ്
- **Streaming Processing**: മെമ്മറി പ്രശ്നങ്ങൾ ഒഴിവാക്കി വലിയ ഡോക്യുമെന്റുകൾ കൈകാര്യം ചെയ്യുക
- **Resource Pooling**: ചെലവേറിയ റിസോഴ്‌സുകളുടെ കാര്യക്ഷമമായ പുനരുപയോഗം
- **Garbage Collection**: മെമ്മറി വിന്യാസത്തിനുള്ള ഒപ്റ്റിമൈസ്ഡ് പാറ്റേണുകൾ

### കാഷിംഗ് നയങ്ങൾ

- **Index Caching**: ഒപ്റ്റിമൈസ്ഡ് വെക്ടർ ഇൻഡക്സ് കാഷിംഗ്
- **Result Caching**: സൃഷ്ടിക്കപ്പെട്ട മറുപടികളുടെ ബുദ്ധിപൂർണ്ണ കാഷിംഗ്

````polyglot-notebook
using System;
using System.Linq;
using Azure.AI.Agents.Persistent;
using Azure.Identity;
using Microsoft.Agents.AI;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
var azure_foundry_endpoint = Environment.GetEnvironmentVariable("AZURE_AI_PROJECT_ENDPOINT") ?? throw new InvalidOperationException("AZURE_AI_PROJECT_ENDPOINT is not set.");
var azure_foundry_model_id = Environment.GetEnvironmentVariable("AZURE_AI_MODEL_DEPLOYMENT_NAME") ?? "gpt-4.1-mini";
````

````polyglot-notebook
string pdfPath = "./document.md";
````

````polyglot-notebook
using System.IO;

async Task<Stream> OpenImageStreamAsync(string path)
{
	return await Task.Run(() => File.OpenRead(path));
}

var pdfStream = await OpenImageStreamAsync(pdfPath);
````

````polyglot-notebook
var persistentAgentsClient = new PersistentAgentsClient(azure_foundry_endpoint, new AzureCliCredential());
````

````polyglot-notebook
PersistentAgentFileInfo fileInfo = await persistentAgentsClient.Files.UploadFileAsync(pdfStream, PersistentAgentFilePurpose.Agents, "demo.md");
````

````polyglot-notebook
PersistentAgentsVectorStore fileStore =
            await persistentAgentsClient.VectorStores.CreateVectorStoreAsync(
                [fileInfo.Id],
                metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
PersistentAgent agentModel = await persistentAgentsClient.Administration.CreateAgentAsync(
            azure_foundry_model_id,
            name: "DotNetRAGAgent",
            tools: [new FileSearchToolDefinition()],
            instructions: """
                You are an AI assistant designed to answer user questions using only the information retrieved from the provided document(s).

                - If a user's question cannot be answered using the retrieved context, **you must clearly respond**: 
                "I'm sorry, but the uploaded document does not contain the necessary information to answer that question."
                - Do not answer from general knowledge or reasoning. Do not make assumptions or generate hypothetical explanations.
                - Do not provide definitions, tutorials, or commentary that is not explicitly grounded in the content of the uploaded file(s).
                - If a user asks a question like "What is a Neural Network?", and this is not discussed in the uploaded document, respond as instructed above.
                - For questions that do have relevant content in the document (e.g., Contoso's travel insurance coverage), respond accurately, and cite the document explicitly.

                You must behave as if you have no external knowledge beyond what is retrieved from the uploaded document.
                """,
            toolResources: new()
            {
                FileSearch = new()
                {
                    VectorStoreIds = { fileStore.Id },
                }
            },
            metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
AIAgent agent = await persistentAgentsClient.GetAIAgentAsync(agentModel.Id);
````

````polyglot-notebook
AgentThread thread = agent.GetNewThread();
````

````polyglot-notebook
Console.WriteLine(await agent.RunAsync("Can you explain Contoso's travel insurance coverage?", thread));
````
