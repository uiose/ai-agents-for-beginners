## പടി 1 - ഒരു ചെറിയ ഉദാഹരണത്തോടെ ആരംഭിക്കുക

അടുത്ത സെൽ എന്താണ് ചെയ്യുന്നത് എന്ന് ലളിതമായ ഭാഷയിൽ وض്ഘീകരിക്കുക


```python
# Minimal working example
angles = [0, math.pi / 4, math.pi / 2]
sines = [math.sin(a) for a in angles]
list(zip(angles, sines))
```

## അഭ്യാസങ്ങൾ

- ഒരു സാധാരണപ്പെട്ട പിശക് ഒന്ന് രേഖപ്പെടുത്തുക, അത് എങ്ങനെ പരിഹരിക്കാമെന്നും പറയുക.


```python
# Exercise answer scaffold
def describe(values: list[float]) -> dict[str, float]:
    return {"min": min(values), "max": max(values)}

describe(sines)
```
