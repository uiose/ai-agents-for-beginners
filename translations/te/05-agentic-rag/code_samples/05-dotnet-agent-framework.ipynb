# 🔍 ఎంటర్ప్రైజ్ RAG Microsoft Foundry (.NET)తో

## 📋 అభ్యాసన లక్ష్యాలు

ఈ నోట్‌బుక్ Microsoft Agent Framework ను .NETలో Microsoft Foundryతో ఉపయోగించి ఎంటర్ప్రైజ్-గ్రేడ్ Retrieval-Augmented Generation (RAG) వ్యవస్థలను ఎలా నిర్మించాలో చూపిస్తుంది. మీరు డాక్యుమెంట్లలో శోధించి ఎంటర్ప్రైజ్ భద్రత మరియు స్కేలబిలిటీతో ఖచ్చితమైన, సందర్భానుసారమైన జవాబులను అందించే ప్రొడక్షన్-రెడీ ఏజెంట్లను సృష్టించడం నేర్చుకుంటారు.

**మీరు నిర్మించబోతున్న ఎంటర్ప్రైజ్ RAG సామర్ధ్యాలు:**
- 📚 **డాక్యుమెంట్ ఇంటెలిజెన్స్**: Azure AI సేవలతో అధునాతన డాక్యుమెంట్ ప్రాసెసింగ్
- 🔍 **సెమాంటిక్ శోధన**: ఎంటర్ప్రైజ్ ఫీచర్లతో ఉన్న అధిక-గతిత్వ వెక్టర్ శోధన
- 🛡️ **భద్రత సమీకరణం**: పాత్ర-ఆధారిత ప్రాప్తి మరియు డేటా రక్షణ నమూనాలు
- 🏢 **స్కేలబుల్ ఆర్కిటెక్చర్**: మానిటరింగ్‌తో ప్రొడక్షన్-రెడీ RAG వ్యవస్థలు

## 🎯 ఎంటర్ప్రైజ్ RAG ఆర్కిటెక్చర్

### కోర్ ఎంటర్ప్రైజ్ భాగాలు
- **Microsoft Foundry**: భద్రత మరియు కంప్లయన్స్‌తో మేనేజ్ చేయబడే ఎంటర్ప్రైజ్ AI ప్లాట్‌ఫారమ్
- **Persistent Agents**: సంభాషణ చరిత్ర మరియు సందర్భ నిర్వహణతో స్టేట్‌ఫుల్ ఏజెంట్లు
- **Vector Store Management**: ఎంటర్ప్రైజ్-గ్రేడ్ డాక్యుమెంట్ ఇండెక్సింగ్ మరియు రిట్రీవల్
- **Identity Integration**: Azure AD ఆథెంటికేషన్ మరియు పాత్ర-ఆధారిత ప్రాప్తి నియంత్రణ

### .NET ఎంటర్ప్రైజ్ లాభాలు
- **Type Safety**: RAG ఆపరేషన్లు మరియు డేటా స్ట్రక్చర్ల కోసం కంపైల్-టైమ్ ధృవీకరణ
- **Async Performance**: అప్రతిబంధక డాక్యుమెంట్ ప్రాసెసింగ్ మరియు శోధన ఆపరేషన్లు
- **Memory Management**: పెద్ద డాక్యుమెంట్ సేకరణల కోసం సమర్థవంతమైన రిసోర్స్ వినియోగం
- **Integration Patterns**: డిపెండెన్సీ ఇంజక్షన్‌తో సహజ Azure సేవ సమీకరణ

## 📊 ఎంటర్ప్రైజ్ RAG నమూనాలు

### డాక్యుమెంట్ మేనేజ్‌మెంట్ నమూనాలు
- **Bulk Upload**: పెద్ద డాక్యుమెంట్ సేకరణల యొక్క సమర్థవంతమైన ప్రాసెసింగ్
- **Incremental Updates**: రియల్-టైమ్ డాక్యుమెంట్ చేర్పు మరియు సవరణ
- **Version Control**: డాక్యుమెంట్ వెర్షనింగ్ మరియు మార్పుల ట్రాకింగ్
- **Metadata Management**: సమృద్ధి డాక్యుమెంట్ లక్షణాలు మరియు టాక్సొనమీ

### శోధన & రిట్రీవల్ నమూనాలు
- **Hybrid Search**: ఉత్తమ ఫలితాల కోసం సెమాంటిక్ మరియు కీవర్డ్ శోధన కలయిక
- **Faceted Search**: బహు-పరిమాణ ఫిల్టరింగ్ మరియు వర్గీకరణ
- **Relevance Tuning**: డొమైన్-స్పెసిఫిక్ అవసరాల కోసం కస్టమ్ స్కోరింగ్ అల్గోరిథంలు
- **Result Ranking**: బిజినెస్ లాజిక్ సమీకరణతో అధునాతన ర్యాంకింగ్

### భద్రత నమూనాలు
- **Document-Level Security**: డాక్యుమెంటుకు ప్రతీదానికీ సూక్ష్మ-గ్రాన్యులర్ ప్రాప్తి నియంత్రణ
- **Data Classification**: ఆటోమేటిక్ సెన్సిటివిటీ లేబలింగ్ మరియు రక్షణ
- **Audit Trails**: అన్ని RAG ఆపరేషన్ల యొక్క సమగ్ర లాగింగ్
- **Privacy Protection**: PII గుర్తింపు మరియు రెడాక్షన్ సామర్థ్యాలు

````polyglot-notebook
using System;
using System.Linq;
using Azure.AI.Agents.Persistent;
using Azure.Identity;
using Microsoft.Agents.AI;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
var azure_foundry_endpoint = Environment.GetEnvironmentVariable("AZURE_AI_PROJECT_ENDPOINT") ?? throw new InvalidOperationException("AZURE_AI_PROJECT_ENDPOINT is not set.");
var azure_foundry_model_id = Environment.GetEnvironmentVariable("AZURE_AI_MODEL_DEPLOYMENT_NAME") ?? "gpt-4.1-mini";
````

````polyglot-notebook
string pdfPath = "./document.md";
````

````polyglot-notebook
using System.IO;

async Task<Stream> OpenImageStreamAsync(string path)
{
	return await Task.Run(() => File.OpenRead(path));
}

var pdfStream = await OpenImageStreamAsync(pdfPath);
````

````polyglot-notebook
var persistentAgentsClient = new PersistentAgentsClient(azure_foundry_endpoint, new AzureCliCredential());
````

````polyglot-notebook
PersistentAgentFileInfo fileInfo = await persistentAgentsClient.Files.UploadFileAsync(pdfStream, PersistentAgentFilePurpose.Agents, "demo.md");
````

````polyglot-notebook
PersistentAgentsVectorStore fileStore =
            await persistentAgentsClient.VectorStores.CreateVectorStoreAsync(
                [fileInfo.Id],
                metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
PersistentAgent agentModel = await persistentAgentsClient.Administration.CreateAgentAsync(
            azure_foundry_model_id,
            name: "DotNetRAGAgent",
            tools: [new FileSearchToolDefinition()],
            instructions: """
                You are an AI assistant designed to answer user questions using only the information retrieved from the provided document(s).

                - If a user's question cannot be answered using the retrieved context, **you must clearly respond**: 
                "I'm sorry, but the uploaded document does not contain the necessary information to answer that question."
                - Do not answer from general knowledge or reasoning. Do not make assumptions or generate hypothetical explanations.
                - Do not provide definitions, tutorials, or commentary that is not explicitly grounded in the content of the uploaded file(s).
                - If a user asks a question like "What is a Neural Network?", and this is not discussed in the uploaded document, respond as instructed above.
                - For questions that do have relevant content in the document (e.g., Contoso's travel insurance coverage), respond accurately, and cite the document explicitly.

                You must behave as if you have no external knowledge beyond what is retrieved from the uploaded document.
                """,
            toolResources: new()
            {
                FileSearch = new()
                {
                    VectorStoreIds = { fileStore.Id },
                }
            },
            metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
AIAgent agent = await persistentAgentsClient.GetAIAgentAsync(agentModel.Id);
````

````polyglot-notebook
AgentThread thread = agent.GetNewThread();
````

````polyglot-notebook
Console.WriteLine(await agent.RunAsync("Can you explain Contoso's travel insurance coverage?", thread));
````
