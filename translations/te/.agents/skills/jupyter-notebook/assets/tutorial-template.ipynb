## దశ 1 - చిన్న ఉదాహరణతో ప్రారంభించండి

తదుపరి సెల్ ఏమి చేస్తుందో సాధారణ భాషలో వివరించండి.


```python
# Minimal working example
angles = [0, math.pi / 4, math.pi / 2]
sines = [math.sin(a) for a in angles]
list(zip(angles, sines))
```

## అభ్యాసాలు

- వేరే ఇన్‌పుట్ ప్రయత్నించండి.
- కోడ్‌ను నడిపే ముందు ఫలితాన్ని అంచనా వేయండి.
- ఒక సాధారణ పొరపాటు మరియు దాన్ని ఎలా సరిచేయాలో గమనించండి.


```python
# Exercise answer scaffold
def describe(values: list[float]) -> dict[str, float]:
    return {"min": min(values), "max": max(values)}

describe(sines)
```
