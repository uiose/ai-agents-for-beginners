# Lesson 01 - AI ਏਜੰਟਾਂ ਦਾ ਪਰਿਚਯ

**ਸ਼ੁਰੂਆਤੀ ਲੋਕਾਂ ਲਈ AI ਏਜੰਟਾਂ** ਕੋਰਸ ਵਿੱਚ ਤੁਹਾਡਾ ਸਵਾਗਤ ਹੈ!

ਇੱਕ **AI ਏਜੰਟ** ਇੱਕ ਪ੍ਰੋਗਰਾਮ ਹੁੰਦਾ ਹੈ ਜੋ ਵੱਡੇ ਭਾਸ਼ਾ ਮਾਡਲ (LLM) ਨੂੰ ਆਪਣੇ ਤਰਕਸ਼ੀਲ ਇੰਜਣ ਵਜੋਂ ਵਰਤਦਾ ਹੈ ਅਤੇ ਅਸਲੀ ਦੁਨੀਆ ਵਿੱਚ *ਕਿਰਦਾਰ* ਨਿਭਾ ਸਕਦਾ ਹੈ — API ਕਾਲ ਕਰਨਾ, ਡੇਟਾਬੇਸ ਪੁੱਛ-ਗਿੱਛ ਕਰਨਾ, ਜਾਂ ਕੋਡ ਚਲਾਉਣਾ — ਯੂਜ਼ਰ ਦੀ ਓਰੋਂ ਕਿਸੇ ਲਕਸ਼ ਨੂੰ ਹਾਸਲ ਕਰਨ ਲਈ।

ਇਸ ਨੋਟਬੁੱਕ ਵਿੱਚ ਤੁਸੀਂ ਆਪਣਾ ਪਹਿਲਾ ਏਜੰਟ ਬਣਾਉਗੇ: ਇੱਕ **ਟ੍ਰੈਵਲ ਏਜੰਟ** ਜੋ ਰਹਿਣ ਵਾਲੀਆਂ ਜਗ੍ਹਾ ਦੀ ਸਿਫਾਰਿਸ਼ ਕਰਦਾ ਹੈ। ਇਸ ਦੌਰਾਨ ਤੁਸੀਂ ਸਿੱਖੋਗੇ ਕਿ:

1. **ਮਾਈਕ੍ਰੋਸੌਫਟ ਏਜੰਟ ਫਰੇਮਵਰਕ** ਦੀ ਵਰਤੋਂ ਕਰਕੇ Azure AI Foundry Agent Service ਨਾਲ ਕਿਵੇਂ ਜੁੜਨਾ ਹੈ।
2. ਏਜੰਟ ਨੂੰ ਇੱਕ **ਟੂਲ** ਦੇਣਾ — ਇੱਕ ਸਿੱਧੀ Python ਫੰਕਸ਼ਨ ਜੋ ਉਹ ਕਾਲ ਕਰ ਸਕਦਾ ਹੈ।
3. ਏਜੰਟ ਨੂੰ ਰਨ ਕਰਨਾ ਅਤੇ ਇਸ ਦੇ ਜਵਾਬ ਦਾ ਨਿਰੀਖਣ ਕਰਨਾ।
4. ਏਜੰਟ ਦੇ ਜਵਾਬ ਨੂੰ ਟੋਕਨ-ਦਰ-ਟੋਕਨ ਸਟ੍ਰੀਮ ਕਰਨਾ।


## ਸੈਟਅਪ

ਇਸ ਨੋਟਬੁੱਕ ਨੂੰ ਚਲਾਉਣ ਤੋਂ ਪਹਿਲਾਂ, ਇਹ ਯਕੀਨੀ ਬਣਾਓ ਕਿ ਤੁਹਾਡੇ ਕੋਲ ਹੈ:

1. **ਇੱਕ Azure AI Foundry ਪ੍ਰੋਜੈਕਟ** ਜਿਸ ਨਾਲ ਇੱਕ ਚੈਟ ਮਾਡਲ ਤੈਨਾਤ ਕੀਤਾ ਗਿਆ ਹੈ (ਜਿਵੇਂ ਕਿ `gpt-4o-mini`)।
2. **Azure CLI ਨਾਲ ਲੋਗਇਨ ਕੀਤਾ ਹੈ** — ਆਪਣੇ ਟਰਮੀਨਲ ਵਿੱਚ `az login` ਚਲਾਓ।
3. **ਲੋੜੀਂਦੇ ਵਾਤਾਵਰਣ ਚੱਲਣ ਵਾਲੇ ਤੱਤ ਸੈੱਟ ਕੀਤੇ ਹਨ:**
   - `AZURE_AI_PROJECT_ENDPOINT` — ਤੁਹਾਡੇ Azure AI Foundry ਪ੍ਰੋਜੈਕਟ ਦਾ ਐਂਡਪੁਆਇੰਟ।
   - `AZURE_AI_MODEL_DEPLOYMENT_NAME` — ਤੁਹਾਡੇ ਤੈਨਾਤ ਮਾਡਲ ਦਾ ਨਾਮ।

ਹੇਠਾਂ ਦਿੱਤਾ ਸੈੱਲ ਤੁਹਾਨੂੰ ਲੋੜੀਂਦੇ Python ਪੈਕੇਜ ਇੰਸਟਾਲ ਕਰਦਾ ਹੈ।

```python
import logging
logging.getLogger("agent_framework.azure").setLevel(logging.ERROR)

import os
import asyncio
from typing import Annotated

from agent_framework import tool
from agent_framework.azure import AzureAIProjectAgentProvider
from azure.identity import AzureCliCredential

provider = AzureAIProjectAgentProvider(credential=AzureCliCredential())
```

## ਆਪਣੇ ਪਹਿਲੇ ਏਜੰਟ ਨੂੰ ਬਣਾਉਣਾ

ਏਜੰਟ ਨੂੰ ਦੋ ਚੀਜ਼ਾਂ ਦੀ ਲੋੜ ਹੁੰਦੀ ਹੈ:

- **ਹਦਾਇਤਾਂ** ਜੋ ਇਹ ਦੱਸਦੀਆਂ ਹਨ ਕਿ *ਉਹ ਕੌਣ ਹੈ* ਅਤੇ *ਕਿਵੇਂ ਵਤੀਰਾ ਕਰਨਾ ਹੈ* (ਇੱਕ ਸਿਸਟਮ ਪ੍ਰਾਂਪਟ)।
- **ਟੂਲ** — Python ਫੰਕਸ਼ਨਾਂ ਨੂੰ `@tool` ਨਾਲ ਸਜਾਇਆ ਗਿਆ ਜੋ ਏਜੰਟ ਜਾਣਕਾਰੀ ਪ੍ਰਾਪਤ ਕਰਨ ਜਾਂ ਕਾਰਵਾਈ ਕਰਨ ਲਈ ਕਾਲ ਕਰ ਸਕਦਾ ਹੈ।

ਹੇਠਾਂ ਅਸੀਂ ਇੱਕ ਸਧਾਰਣ ਟੂਲ ਨੂੰ ਪਰਿਭਾਸ਼ਿਤ ਕਰਦੇ ਹਾਂ ਜੋ ਲੋਕਪ੍ਰਿਯ ਛੁੱਟੀਆਂ ਵਾਲੀਆਂ ਜਗ੍ਹਾਂ ਦੀ ਸੂਚੀ ਵਾਪਸ ਕਰਦਾ ਹੈ। ਜਦੋਂ ਯੂਜ਼ਰ ਯਾਤਰਾ ਦੀ ਸਿਫਾਰਸ਼ਾਂ ਮੰਗਦਾ ਹੈ ਤਾਂ ਏਜੰਟ ਇਸ ਟੂਲ ਦਾ ਉਪਯੋਗ ਕਰੇਗਾ।


```python
@tool(approval_mode="never_require")
def get_destinations() -> list[str]:
    """Get a list of popular vacation destinations."""
    return [
        "Barcelona",
        "Paris",
        "Berlin",
        "Tokyo",
        "Sydney",
        "New York City",
        "Cairo",
        "Cape Town",
        "Rio de Janeiro",
        "Bali",
    ]
```

```python
agent = await provider.create_agent(
    tools=[get_destinations],
    name="TravelAgent",
    instructions=(
        "You are a helpful travel agent. Help users find their perfect vacation "
        "destination based on their preferences. Use the get_destinations tool "
        "to see available destinations."
    ),
)

response = await agent.run(
    "I'm looking for a warm beach destination. What do you recommend?"
)
print(response)
```

## Streaming Responses

ਇੱਕ ਵਧੇਰੇ ਅੰਤਰਕਿਰਿਆ ਪਰਦਾਨ ਕਰਨ ਲਈ ਤੁਸੀਂ ਏਜੰਟ ਦੇ ਜਵਾਬ ਨੂੰ **ਸਟ੍ਰੀਮ** ਕਰ ਸਕਦੇ ਹੋ। ਪੂਰੇ ਜਵਾਬ ਦੀ ਉਡੀਕ ਕਰਨ ਦੀ ਬਜਾਏ, ਏਜੰਟ ਟੈਕਸਟ ਦੇ ਟੁਕੜੇ ਜੌ ਦੇ ਰੂਪ ਵਿੱਚ ਜਾਰੀ ਕਰਦਾ ਹੈ ਜਿਵੇਂ ਉਹ ਬਣਾਏ ਜਾਂਦੇ ਹਨ। ਇਹ ਖਾਸ ਕਰਕੇ ਚੈਟ ਇੰਟਰਫੇਸਾਂ ਵਿੱਚ ਲਾਭਦਾਇਕ ਹੈ ਜਿੱਥੇ ਤੁਸੀਂ ਨਤੀਜੇ ਨੂੰ ਅਸਲ ਸਮੇਂ ਵਿੱਚ ਦਿਖਾਉਣਾ ਚਾਹੁੰਦੇ ਹੋ।


```python
async for chunk in agent.run(
    "Tell me about Tokyo as a travel destination", stream=True
):
    print(chunk, end="", flush=True)
```
