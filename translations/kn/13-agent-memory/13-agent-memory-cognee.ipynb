ಈ ನೋಟ್ಬುಕ್ ಬುದ್ಧಿವಂತ **ಕೋಡಿಂಗ್ ಸಹಾಯಕ** ಅನ್ನು ಸ್ಥಿರ ಸ್ಮೃತಿಯೊಂದಿಗೆ ಹೇಗೆ ನಿರ್ಮಿಸಬೇಕು ಎಂಬುದನ್ನು [**Cognee**](https://www.cognee.ai/) ಜ್ಞಾನ ಗ್ರಾಫ್‌ಗಳು ಮತ್ತು **Microsoft Agent Framework** (MAF) ಬಳಸಿ ತೋರಿಸುತ್ತದೆ.

Cognee ರಚನಾತ್ಮಕವಲ್ಲದ ಪಠ್ಯವನ್ನು ರಚನಾತ್ಮಕ, ಪ್ರಶ್ನೆ ಮಾಡಬಹುದಾದ ಜ್ಞಾನ ಗ್ರಾಫ್‌ಗೆ ಪರಿವರ್ತಿಸುತ್ತದೆ, ಇದು ವೆಕ್ಟರ್ ಎम्बೆಡ್ಡಿಂಗ್ಸ್ ಮೂಲಕ ಬೆಂಬಲಿತವಾಗಿರುತ್ತದೆ — ನಿಮ್ಮ ಏಜಂಟ್‌ಗೆ ಸಮೃದ್ಧ, ಸಂಬಂಧ-ಅನುವೇದ್ಯ ಧೀರ್ಘಕಾಲಿಕ ಸ್ಮೃತಿಯನ್ನು ನೀಡುತ್ತದೆ.

### ನೀವು ಏನು ಕಲಿಯುತ್ತೀರಿ
1. **ಜ್ಞಾನ ಗ್ರಾಫ್‌ಗಳ ನಿರ್ಮಾಣ**: ಡೆವಲಪರ್ ಪ್ರೊಫೈಲ್‌ಗಳು ಮತ್ತು ಉತ್ತಮ ಅಭ್ಯಾಸಗಳನ್ನು ರಚನಾತ್ಮಕ, ಪ್ರಶ್ನೆ ಮಾಡಬಹುದಾದ ಜ್ಞಾನದಲ್ಲಿ ಪರಿವರ್ತಿಸುವುದು.
2. **Cognee ಅನ್ನು MAF ಜತೆ ಸಂಯೋಜಿಸುವುದು**: `@tool` ಫಂಕ್ಷನ್ಗಳನ್ನು ಬಳಸಿಕೊಂಡು MAF ಏಜಂಟ್ Cognee ನ ಜ್ಞಾನ ಗ್ರಾಫ್ ಅನ್ನು ಪ್ರಶ್ನಿಸಬಹುದಾಗಿಸುವುದು.
3. **ಸೆಷನ್-ಅನುವೇದ್ಯ ಸಂಭಾಷಣೆಗಳು**: ಒಂದೇ ಸೆಷನ್‌ನಲ್ಲಿ ಅನೇಕ ಪ್ರಶ್ನೆಗಳ ನಡುವೆ ಸಂದರ್ಭವನ್ನು ಕಾಪಾಡುವುದು.
4. **ಧೀರ್ಘಕಾಲಿಕ ಸ್ಮೃತಿ**: ಸೆಷನ್‌ಗಳ ನಡುವಿನ ಪ್ರಮುಖ ಜ್ಞಾನವನ್ನು ಸ್ಥಿರವಾಗಿರಿಸುವುದು ಮತ್ತು ಹೊಸ ಸಂಭಾಷಣೆಗಳಲ್ಲಿ ಅದನ್ನು ಪಡೆದುಕೊಳ್ಳುವುದು.

### Cognee ಯ ಮೆಮೊರಿ ರಚನೆ
```
┌──────────────────────────┐
│      Raw Data            │  (developer profiles, docs, conversations)
└───────────┬──────────────┘
            │  cognee.add() + cognee.cognify()
            ▼
┌──────────────────────────────────────────┐
│  Knowledge Graph + Vector Embeddings     │
└───────────┬──────────────────────────────┘
            │  cognee.search()
            ▼
┌──────────────────┐       ┌────────────────┐
│  MAF Agent       │──────▶│  @tool funcs   │
│  (AgentSession)  │       │  wrapping       │
│                  │       │  cognee.search  │
└──────────────────┘       └────────────────┘
```

## ಭಾಗ 2 — Cognee ಉಪಕರಣಗಳೊಂದಿಗೆ MAF ಏಜೆಂಟ್

ಈಗ ನಾವು `@tool` ಫಂಕ್ಷನ್‌ಗಳ ಮೂಲಕ Cognee ಅವರ ಜ್ಞಾನ ಗ್ರಾಫ್ ಅನ್ನು ಕ್ವೇರಿ ಮಾಡಬೇಕಾದ MAF ಏಜೆಂಟ್ ಅನ್ನು ರಚಿಸುತ್ತೇವೆ. ಇದರಿಂದ ಏಜೆಂಟ್ ಗ್ರಾಫ್ ತಿಳುವಳಿಕೆಯ ಸ್ಯಾಮೆಂಟಿಕ್ ಶೋಧನೆಯನ್ನು ಸಂಪೂರ್ಣವಾಗಿ ಬಳಸಿಕೊಂಡು, ಸೆಷನ್‌ಗಳ ಮೂಲಕ ಸಂಭಾಷಣಾತ್ಮಕ ಕುರಿತು ಕಾಯ್ದುಕೊಳ್ಳಲು ಅವಕಾಶ ಸಿಗುತ್ತದೆ.


````python
@tool(approval_mode="never_require")
async def search_knowledge(
    query: Annotated[str, "Natural-language question to search the knowledge graph"],
) -> str:
    """Search the Cognee knowledge graph for relevant developer knowledge, best practices, and past conversations."""
    results = await cognee.search(
        query_text=query,
        query_type=SearchType.GRAPH_COMPLETION,
    )
    if not results:
        return "No relevant knowledge found."
    return str(results)


@tool(approval_mode="never_require")
async def search_principles(
    query: Annotated[str, "Question about Python principles or best practices"],
) -> str:
    """Search only the Python principles subset of the knowledge graph."""
    from cognee.modules.engine.models.node_set import NodeSet
    results = await cognee.search(
        query_text=query,
        query_type=SearchType.GRAPH_COMPLETION,
        node_type=NodeSet,
        node_name=["principles_data"],
    )
    if not results:
        return "No relevant principles found."
    return str(results)


print("✅ Cognee tools defined: search_knowledge, search_principles")
````

````python
coding_agent = await provider.create_agent(
    name="CodingAssistant",
    instructions=(
        "You are an expert coding assistant with access to a knowledge graph "
        "containing developer profiles, Python best practices, and past conversations.\n\n"
        "WORKFLOW:\n"
        "1. Use search_knowledge() to find relevant information from the full knowledge graph.\n"
        "2. Use search_principles() when the question is specifically about Python best practices.\n"
        "3. Combine retrieved knowledge with your own expertise to give comprehensive answers.\n"
        "4. Reference the developer's known tech stack (FastAPI, asyncio, Pydantic) when relevant."
    ),
)

print("✅ CodingAssistant agent created")
````

## ಸೆಷನ್ಗಳೊಂದಿಗೆ ಕಾರ್ಯೋನ್ಮುಖ ಸ್ಮೃತಿ

`AgentSession` (`agent.create_session()` ಮೂಲಕ ರಚಿಸಲಾಗಿದೆ) ಸೆಷನ್ ಒಳಗಿನ ಕಾರ್ಯೋನ್ಮುಖ ಸ್ಮೃತಿಯನ್ನು ಒದಗಿಸುತ್ತದೆ. ಏಜೆಂಟ್ ಹಿಂದಿನ ಸಂದೇಶಗಳಿಗೆ ಹಿಂತಿರುಗಿ ಗಮನಿಸಬಹುದು ಮತ್ತು ಅದೇ ಸಮಯದಲ್ಲಿ Cognee ನ ದೀರ್ಘಕಾಲೀನ ಜ್ಞಾನ ಗ್ರಾಫ್ ಅನ್ನು ಪ್ರಶ್ನಿಸಬಹುದು.


````python
session = coding_agent.create_session()

response = await coding_agent.run(
    "How does my AsyncWebScraper implementation align with Python's design principles?",
    session=session,
)
print("🤖 Agent:", response)
````

````python
response = await coding_agent.run(
    "Based on what you just said, when should I pick dataclasses versus Pydantic for this work?",
    session=session,
)
print("🤖 Agent:", response)
print("\n💡 The agent combined working memory (previous answer) with Cognee's knowledge graph.")
````

## ಸಾರಾಂಶ

ಈ ನೋಟ್ಬುಕ್‌ನಲ್ಲಿ ನೀವು **MAF ನ ಕೆಲಸದ ಸ್ಮರಣೆ** (`agent.create_session()`) ಮತ್ತು **Cognee ನ ದೀರ್ಘಕಾಲीन ಜ್ಞಾನ ಗ್ರಾಫ್** ಅನ್ನು ಸಂಯೋಜಿಸುವ ಕೋಡಿಂಗ್ ಸಹಾಯಕನನ್ನ ನಿರ್ಮಿಸಿದ್ದೀರಾ.

### ನೀವು ಕಲಿತದ್ದೇನು
1. **ಜ್ಞಾನ ಗ್ರಾಫ್ ನಿರ್ಮಾಣ**: Cognee ಅಸಂರಚಿತ ಪಠ್ಯವನ್ನು ಸ್ವೀಕರಿಸಿ ಗ್ರಾಫ್ + ವೆಕ್ಟರ್ ಸ್ಮರಣೆಯನ್ನು ನಿರ್ಮಿಸುತ್ತದೆ.
2. **memify ನೊಂದಿಗೆ ಗ್ರಾಫ್ ಅಭಿವೃದ್ದಿ**: ನಿಮ್ಮ ಇ существующий ಗ್ರಾಫ್ ಮೇರೆಗೆ ನಿರ ಸಂಪರ್ಕಿತ ನಿಯಮಗಳು ಮತ್ತು ಸುದೀರ್ಘ ಸಂಬಂಧಗಳನ್ನು ಸೃಷ್ಟಿಸುತ್ತದೆ.
3. **MAF + Cognee ಸಂಯೋಜನೆ**: `@tool` ಫಂಕ್ಷನ್‌ಗಳು MAF ಏಜೆಂಟ್ ಗೆ Cognee ನ ಗ್ರಾಫ್ ಅನ್ನು ಸಹಜವಾಗಿ ಪ್ರಶ್ನೆ ಮಾಡುವುದಕ್ಕೆ ಅವಕಾಶ ಮಾಡಿಕೊಡುತ್ತವೆ.
4. **ಕೆಲಸದ ಮೊದಲು ಅರ್ಧ ಸ್ಮರಣೆ + ದೀರ್ಘಕಾಲीन ಸ್ಮರಣೆ**: `AgentSession` (`agent.create_session()`) ಸೆಷನ್ ಸಂಧರ್ಭವನ್ನು ಒದಗಿಸುತ್ತದೆ, Cognee ಸ್ಥಾಯಿ ಜ್ಞಾನ ಒದಗಿಸುತ್ತದೆ.
5. **NodeSets ನೊಂದಿಗೆ ಫಿಲ್ಟರಿಂಗ್ ಸಂಶೋಧನೆ**: ಜ್ಞಾನ ಗ್ರಾಫ್ ನ ಕೆಲವು ನಿರ್ದಿಷ್ಟ ಉಪಗ್ರೂಪಗಳನ್ನು ಗುರಿಯಾಗಿಸಬಹುದು (ಉದಾ. ನಿಯಮಗಳು ಮಾತ್ರ).

### ಮುಖ್ಯ ಅಂಶಗಳು
- **Cognee** ಅಮೂರ್ತ ಪಠ್ಯವನ್ನು ರಚನಾತ್ಮಕ, ಸಂಬಂಧ-ಜಾಗರೂಕ ಸ್ಮರಣೆಯಾಗಿ ಪರಿವರ್ತಿಸುತ್ತದೆ — ಸರಳ ವೆಕ್ಟರ್ ಸಂಗ್ರಹಕ್ಕಿಂತ ಗಟ್ಟಿಯಾಗಿರುವುದು.
- **`@tool` ಫಂಕ್ಷನ್‌ಗಳು** MAF ಏಜೆಂಟ್‌ಗಳು ಮತ್ತು ಹೊರಗಿನ ಜ್ಞಾನ ವ್ಯವಸ್ಥೆಗಳ ನಡುವೆ ಪಳ್ಳಿ ಸೇತುವೆಯಂತೆ ಕೆಲಸ ಮಾಡುತ್ತವೆ.
- **`AgentSession`** (`agent.create_session()`) ಪ್ರತಿ ಸಂಭಾಷಣೆಯ ಸಂಧರ್ಭವನ್ನು ದೀರ್ಘಕಾಲದ ಜ್ಞಾನದಿಂದ ಪ್ರತ್ಯೇಕವಾಗಿ ಉಳಿಸುತ್ತದೆ.
- ಇದೇ ಜ್ಞಾನ ಗ್ರಾಫ್ ಹಲವು ಸೆಷನ್‌ಗಳಿಗೆ ಮತ್ತು ಏಜೆಂಟ್‌ಗಳಿಗೆ ಸೇವೆ ನೀಡುತ್ತದೆ.

### ನಿಜಜೀವನದ ಬಳಕೆ
- **ಡೆವಲಪರ್ ಸಹಾಯಕರು**: ಕೋಡ್ ವಿಮರ್ಶೆ, ಘಟನೆಯ ವಿಶ್ಲೇಷಣೆ, ಸಂರಚನಾ ಸಹಾಯಕರು
- **ಗ್ರಾಹಕಮುಖಿ ಸಹಾಯಕರು**: ಉತ್ಪನ್ನ ಡಾಕ್‌ಗಳು,FAQಗಳು ಮತ್ತು CRM ಟಿಪ್ಪಣಿಗಳ ಮೇಲೆ ಬೆಂಬಲ ಏಜೆಂಟ್‌ಗಳು
- **ಆಂತರಿಕ ತಜ್ಞ ಸಹಾಯಕರು**: ನೀತಿ, ಕಾನೂನು ಅಥವಾ ಭದ್ರತಾ ಸಹಾಯಕರು ಮಾರ್ಗಸೂಚಿಗಳ ಮೇರೆಗೆ ತರ್ಕಮಾಡುತ್ತಾರೆ
- **ಏಕೀಕೃತ ಡೇಟಾ ಪದರಗಳು**: ರಚನಾತ್ಮಕ ಮತ್ತು ಅಸಂರಚಿತ ಡೇಟಾವನ್ನು ಒಂದೇ ಒಂದು ಪ್ರಶ್ನಿಸಬಹುದಾದ ಗ್ರಾಫ್ ಆಗಿ ಸಂಯೋಜನೆ ಮಾಡುವುದು

### ಮುಂದಿನ ಹಂತಗಳು
- Cognee ಯಲ್ಲಿ ಕಾಲ ಸಂಬಂಧಿತ ಜಾಗೃತಿ ಪ್ರಯೋಗಿಸುವುದು
- ಕ್ಷೇತ್ರ ನಿರ್ದಿಷ್ಟ ಗ್ರಾಫ್ ಗುಣಮಟ್ಟಕ್ಕೆ OWL ಆಂಟೋಲಜಿ ವ್ಯಾಖ್ಯಾನಿಸುವುದು
- ಕಾಲಕ్రమದಲ್ಲಿ ರಿಟ್ರೀವುಲ್ ಸುಧಾರಣೆಗಾಗಿ ಬಳಕೆದಾರ ಪ್ರತಿಕ್ರಿಯಾ ಲೂಪ್ಗಳು ಸೇರಿಸುವುದು
- Cognee ಸ್ಮರಣೆ ಪದರ ಹಂಚಿಕೊಳ್ಳುವ ಬಹು ಏಜೆಂಟ್ ವ್ಯವಸ್ಥೆಗಳಿಗೆ ತಲುಪಿಸುವುದು
