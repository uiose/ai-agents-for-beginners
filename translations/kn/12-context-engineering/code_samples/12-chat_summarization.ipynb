ಈ ನೋಟ್‍ಬುಕ್ ನಡೆದ ಒಳನಡುವುಗಳನ್ನು ನಿರ್ವಹಿಸುವುದಾಗಿ ಮೈಕ್ರೋಸಾಫ್ಟ್ ಏಜೆಂಟ್ ಫ್ರೆ임್ವರ್ಕ್ ಬಳಸಿ ತೋರುತ್ತದೆ. ಸಂಭಾಷಣೆಗಳು ದೊಡ್ಡದಾಗುತ್ತಿರುವಾಗ, ಟೋಕನ್ ಸಂಖ್ಯೆ ಹೆಚ್ಚಾಗುತ್ತದೆ — ಮತ್ತು ಕೊನೆಗೆ ಮಾದರಿಯ ಒಳನಡುವು ವಿಂಡೋವನ್ನು ಮೀರುತ್ತದೆ. ನಾವು ಇದನ್ನು **ಒಳನಡುವು ಸಂಕ್ಷಿಪ್ತಿಕೆ ಮಾದರಿ** ಮತ್ತು **ಏಜೆಂಟ್ ಸ್ಕ್ರಾಚ್‍ಪ್ಯಾಡ್** ಮೂಲಕ ಶಾಶ್ವತ ಸ್ಮರಣಾಶಕ್ತಿಗಾಗಿ ಪರಿಹರಿಸುತ್ತೇವೆ.

1. **ಒಳನಡುವಿನ ನಿರ್ವಹಣೆ ಏಕೆ ಮುಖ್ಯ**: ಟೋಕನ್ ಮಿತಿ ಮತ್ತು ಒಳನಡುವು ವಿಂಡೋಗಳ ಅರ್ಥ
2. **ಒಳನಡುವು ಜ್ಞಾನದ ಏಜೆಂಟ್‌ಗಳು**: ಸ್ವಂತ ಸಂಭಾಷಣೆಯ ಒಳನಡುವನ್ನು ನಿರ್ವಹಿಸುವ ಏಜೆಂಟ್‌ಗಳನ್ನು ನಿರ್ಮಿಸುವುದು
3. **ಒಳನಡುವು ಸಂಕ್ಷಿಪ್ತಿಕೆ ಮಾದರಿ**: ಸಂಭಾಷಣೆಯ ಇತಿಹಾಸವನ್ನು ಸಂಕ್ಷಿಪ್ತಗೊಳಿಸಲು ಉಪಕರಣಗಳನ್ನು ಬಳಸುವುದು

## ಸಂಧರ್ಭ ನಿರ್ವಹಣೆ ಮಹತ್ವದ ಕಾರಣ

ಪ್ರತಿಯೊಬ್ಬ LLM-ನಿಗೂ ಒಂದು ನಿರ್ದಿಷ್ಟ **ಸಂಧರ್ಭ ಕಿಟಕಿ** ಇರುತ್ತದೆ — ಅದು ಒಂದೇ ವಿನಂತಿಯಲ್ಲಿ ಪ್ರಕ್ರಿಯೆ ಮಾಡಲು ಸಾಧ್ಯವಿರುವ ಗರಿಷ್ಠ ಟೋಕನ್ಗಳ ಸಂಖ್ಯೆ. ಬಹು-ತಿರುವಿನ ಸಂವಾದ ಮುಂದುವರಿದಂತೆ:

- **ಟೋಕನ್ ಎಣಿಕೆ ಸರಣಿವಾಗಿ ಹೆಚ್ಚಾಗುತ್ತದೆ** ಪ್ರತಿಯೊಂದು ಬಳಕೆದಾರ ಸಂದೇಶ ಮತ್ತು ಸಹಾಯಕ ಪ್ರತಿಕ್ರಿಯೆಯೊಂದಿಗೆ.
- **ಪ್ರಾಂಪ್ಟ್ ಟೋಕನ್ಗಳು ವೆಚ್ಚವನ್ನು ಆಳಿಸುತ್ತವೆ** ಏಕೆಂದರೆ ಸಂಪೂರ್ಣ ಇತಿಹಾಸವನ್ನು ಪ್ರತಿ ತಿರುವಿನಲ್ಲಿ ಪುನಃ ಕಳುಹಿಸಲಾಗುತ್ತದೆ.
- ಕೊನೆಗೆ ಸಂವಾದವು **ಸಂಧರ್ಭ ಕಿಟಕಿಯನ್ನು ಮೀರುತ್ತದೆ** ಮತ್ತು ಮಾದರಿ either ಕಡಿತಗೊಳ್ಳುತ್ತದೆ ಅಥವಾ ದೋಷ ನೀಡುತ್ತದೆ.

### ಸಂಧರ್ಭ ನಿರ್ವಹಿಸುವ ತಂತ್ರಗಳು

| ತಂತ್ರ | ಇದು ಹೇಗೆ ಕಾರ್ಯನಿರ್ವಹಿಸುತ್ತದೆ | ವ್ಯತಿರೇಕ |
|---|---|---|
| **ಕಡಿತಗೊಳಿಸುವಿಕೆ** | ಹಳೆಯ ಸಂದೇಶಗಳನ್ನು ಬಿಟ್ಟುಬಿಡುವುದು | ಆರಂಭಿಕ ಸಂಧರ್ಭ ಕಳೆದುಹೋಗುತ್ತದೆ |
| **ಸಂಗ್ರಹಣೆ** | ಹಳೆಯ ಸಂದೇಶಗಳನ್ನು ಸಂಗ್ರಹವಾಗಿ ಸಂಕ್ಷಿಪ್ತಗೊಳಿಸುವುದು | ಕೆಲವು ವಿವರಗಳು ಕಳೆದುಹೋಗಬಹುದು, ಆದರೆ ಪ್ರಮುಖ ಅಂಶಗಳು ಫಿಕ್ಸಾಗಿವೆ |
| **Scratchpad / ಹೊರಗಿನ ಸ್ಮರಣೆ** | ಮುಖ್ಯ ವಿಚಾರಗಳನ್ನು ಸಂವಾದದ ಹೊರಗೆ ಸಂಗ್ರಹಿಸುವುದು | ಉಪಕರಣ ಕರೆಯುವಿಕೆಗೆ ಅಗತ್ಯವಿದೆ, ಆದರೆ ಯಾವುದೇ ಕಡಿತ ನಿರ್ವಹಿಸುತ್ತದೆಯಾದರೂ ಉಳಿಯುತ್ತದೆ |

ಈ ನೋಟ್ಬುಕ್‌ನಲ್ಲಿ ನಾವು **ಸಂಗ್ರಹಣೆಯನ್ನೂ** ಮತ್ತು **scratchpad ಸಾಧನ**-ನನ್ನೂ ಸಂಯೋಜಿಸಿ, ಸಂವಾದ ಇತಿಹಾಸ ಸಂಕ್ಷಿಪ್ತಗೊಳ್ಳುವಾಗಲೂ ಏಜೆಂಟ್連 続ತೆಯನ್ನು ಕಾಯ್ದುಕೊಳ್ಳಲು ಸಹಾಯ ಮಾಡುತ್ತೇವೆ.

## ದೀರ್ಘ ಸಂಭಾಷಣೆಯ ಅನುಕರಣೆ

ಸಂದರ್ಭ ಹೇಗೆ ಸಂಗ್ರಹವಾಗುತ್ತದೆ ಎಂಬುದನ್ನು ನೋಡಲು ಬಹು-ತಪಾಲು ಸಂಭಾಷಣೆಯ ಮೂಲಕನಡೆದಿರಿ. ಏಜೆಂಟ್ ಪ್ರಮುಖ ವಿವರಗಳನ್ನು (ಆগ্রಹಗಳು, ಬಜೆಟ್, ಪ್ರಯಾಣದ ದಿನಾಂಕಗಳು) ತಪಾಲುಗಳ ನಡುವೆಯೂ ಉಳಿಸಿಕೊಳ್ಳಬೇಕು ಮತ್ತು ಸತತತೆಯನ್ನು ತೋರಿಸಲು ಅಗತ್ಯವಿದೆ.


```python
session = agent.create_session()

# Turn 1 - Initial preferences
response = await agent.run("I'm planning a trip to Japan. I love sushi, temples, and photography.", session=session)
print(f"Turn 1: {response}\n")

# Turn 2 - More details
response = await agent.run("My budget is $3000 and I'll be traveling solo for 10 days in April.", session=session)
print(f"Turn 2: {response}\n")

# Turn 3 - Test context retention
response = await agent.run("Based on everything I've told you so far, what's the one thing you'd recommend I not miss?", session=session)
print(f"Turn 3: {response}\n")
```

ಏಜೆಂಟ್ ಹಳೆಯ ಮಾತುಕತೆಗಳಿಂದ_CONTEXT ಅನ್ನು ಹೇಗೆ ಕಾಯ್ದುಕೊಳ್ಳುತ್ತದೆ ಎಂದು ಗಮನಿಸಿ — ಅದು ಜಪಾನ್, ಸುಷಿ, ದೇವಾಲಯಗಳು, ಛಾಯಾಗ್ರಹಣ, $3000 ಬಜೆಟ್, ಏಕಾಗಿ ಪ್ರಯಾಣ ಮತ್ತು ಏಪ್ರಿಲ್ ಸಮಯದ ಬಗ್ಗೆ ತಿಳಿದಿದೆ. ಸಣ್ಣ ಸಂಭಾಷಣದಲ್ಲಿ ಇದು ಚೆನ್ನಾಗಿ ಕೆಲಸ ಮಾಡುತ್ತದೆ, ಆದರೆ ಸಂಭಾಷಣೆ ದೊಡ್ಡದಾಯಿಸಿದಂತೆ ಸಂಪೂರ್ಣ ಇತಿಹಾಸವನ್ನು ಮತ್ತೆ ಕಳುಹಿಸುವುದು ದುಬಾರಿಯಾಗುತ್ತದೆ.

ಸಂದರ್ಬ ಸಂಗ್ರಹಣೆಯನ್ನು ನೋಡಲು ಇನ್ನೂ ಕೆಲವು ತಿರುಗು-madu_ ಮೂಲಕ ಸಂಭಾಷಣೆಯನ್ನು ಮುಂದುವರಿಸೋಣ:


```python
# Turn 4 - Expand the conversation
response = await agent.run("What about accommodation? I prefer traditional Japanese inns.", session=session)
print(f"Turn 4: {response}\n")

# Turn 5 - Change of plans
response = await agent.run("Actually, I've changed my mind about the dates. I'll go in October instead for the autumn colors.", session=session)
print(f"Turn 5: {response}\n")

# Turn 6 - Test retention after change
response = await agent.run("Summarize my complete travel plan so far — destination, budget, duration, interests, accommodation, and timing.", session=session)
print(f"Turn 6: {response}\n")
```

## ಸಂಜ್ಞಾನ ಸಂಕ್ಷೇಪಣೆಯ ಮಾದರಿ

ಸಂವಾದ ವೃದ್ಧಿಯಾಗುತ್ತಿರುವಂತೆ, ನಾವು ಸಂಗ್ರಹಿತ ಸಾಂದರ್ಭಿಕ ಮಾಹಿತಿಯನ್ನು ಸಂಕ್ಷಿಪ್ತ ರೂಪದಲ್ಲಿ ಒಗ್ಗೂಡಿಸಲು **ಸಂಕ್ಷೇಪಣಾ ಉಪಕರಣ**ವನ್ನು ಬಳಸಿ सक्छೇವೆ. ಏಜೆಂಟ್ ಈ ಉಪಕರಣವನ್ನು ಮುಖ್ಯ ಇಚ್ಛೆಗಳು ಹಾಗೂ ಆದ್ಯತೆಗಳನ್ನು ದಾಖಲು ಮಾಡಲು ಕರೆಯುತ್ತದೆ ಹೀಗಾಗಿ ಹಳೆಯ ಸಂದೇಶಗಳನ್ನು ತೆಗೆಯಲಾಗಿದೆಯಾದರೂ, ಆವಶ್ಯಕ ಮಾಹಿತಿಯನ್ನು ಉಳಿಸಿಕೊಂಡಿರುತ್ತದೆ.

ಈ ಮಾದರಿ ಹೆಚ್ಚು ಸುಧಾರಿತ ಇತಿಹಾಸ ಕಡಿತಗೊಳಿಸುವಿಕೆಯ ಆಧಾರ:
1. ಏಜೆಂಟ್ ಸಂವಾದದಿಂದ ಪ್ರಮುಖ ವಾಸ್ತವಗಳನ್ನು ಗುರುತಿಸುತ್ತದೆ
2. ಅವುಗಳನ್ನು ಉಳಿಸಲು ಸಂಕ್ಷೇಪಣಾ ಉಪಕರಣವನ್ನು ಕರೆಯುತ್ತದೆ
3. ಹಳೆಯ ಸಂದೇಶಗಳನ್ನು ಸುರಕ್ಷಿತವಾಗಿ ತೆಗೆದುಹಾಕಬಹುದು, ಏಕೆಂದರೆ ಸಂಕ್ಷೇಪಣೆಯು ಮುಖ್ಯ ವಿಷಯಗಳನ್ನು ಹತ್ತಿರದಿಂದ ಹಿಡಿದಿಡುತ್ತದೆ

ಕೆಳಗಿನಲ್ಲಿ ನಾವು `summarize_preferences` ಎಂಬ ಉಪಕರಣವನ್ನು ವ್ಯಾಖ್ಯಾನಿಸುತ್ತೇವೆ, ಏಜೆಂಟ್ ಇದನ್ನು ತನ್ನ ಕಲಿತ ಮುಖ್ಯಾಂಶದ ಸಂಕ್ಷಿಪ್ತ ಸಾರವನ್ನು ದಾಖಲಿಸುವುದಕ್ಕೆ ಕರೆಯಬಹುದು.


```python
@tool(approval_mode="never_require")
def summarize_preferences(conversation_notes: str) -> str:
    """Summarize accumulated user preferences into a compact format."""
    return f"[SUMMARY] User preferences recorded: {conversation_notes}"


# Create an enhanced agent with the summarization tool
summarizing_agent = await provider.create_agent(
    name="SummarizingTravelAgent",
    instructions="""You are a helpful travel planning assistant that actively manages conversation context.

CONTEXT MANAGEMENT RULES:
1. After gathering several user preferences, call summarize_preferences() to record a compact summary
2. When the user asks you to recall details, reference your recorded summaries
3. Keep responses concise — avoid restating the entire history

PLANNING PROCESS:
1. Gather user preferences (destination, budget, dates, interests)
2. Summarize preferences using the tool
3. Create recommendations based on the summary
4. Update the summary when preferences change""",
    tools=[summarize_preferences],
)

print("🤖 Summarizing travel agent created with context tools")
```

```python
# Demonstrate the summarization pattern
summary_session = summarizing_agent.create_session()

# Provide a batch of preferences
response = await summarizing_agent.run(
    "I want to visit Greece. I love seafood, history, and island hopping. "
    "Budget is $4000 for two weeks. Traveling with my partner in June. "
    "Please record these preferences using your summarization tool.",
    session=summary_session,
)
print(f"Agent: {response}\n")

# Ask the agent to use the recorded context
response = await summarizing_agent.run(
    "Now, based on what you've recorded, suggest the top 3 islands we should visit.",
    session=summary_session,
)
print(f"Agent: {response}\n")
```

## ಸಾರಾಂಶ

ಈ ಪಾಠದಲ್ಲಿ ನೀವು ಮೈಕ್ರೋಸಾಫ್ಟ್ ಏಜೆಂಟ್ ಫ್ರೇಮ್‌ವರ್ಕ್ ಬಳಸಿ ದೀರ್ಘಕಾಲ ಚಲಿಸುವ ಏಜೆಂಟ್ ಸಂಭಾಷಣೆಗಳಲ್ಲಿ ಸಂಪರ್ಕವನ್ನು ಹೇಗೆ ನಿರ್ವಹಿಸಬೇಕು ಎಂಬುದನ್ನು ಕಲಿತೀರಿ:

### ಪ್ರಮುಖ ಪರಿಕಲ್ಪನೆಗಳು
- **ಸಂದರ್ಭ ವಿಂಡೋಗಳು ಸೀಮಿತವಾಗಿವೆ** — ಸಂಭಾಷಣೆ ಇತಿಹಾಸದ ಪ್ರತಿಯೊಂದು ಟೋಕನ್‌ಗೆ ಹಣ ಖರ್ಚು ಆಗುತ್ತದೆ ಮತ್ತು ಮಿತಿಗೆ ಸೇರಿಸಲಾಗುತ್ತದೆ.
- **ಸಾರಾಂಶ ಸಾಧನಗಳು** ಏಜೆಂಟ್‌ಗೆ ಸಂಗ್ರಹಿಸಿದ ಸಂಪರ್ಕವನ್ನು ಸಂಕ್ಷಿಪ್ತ ಸಾರಾಂಶಗಳಾಗಿ ಪರಿವರ್ತಿಸುವಂತೆ ಅನುಮತಿಸುತ್ತವೆ, ಇದರಿಂದ ಟೋಕನ್ ಬಳಕೆ ಕಡಿಮೆಯಾಗುತ್ತದೆ ಮತ್ತು ಅಗತ್ಯ ಮಾಹಿತಿಯನ್ನು ಉಳಿಸಲಾಗುತ್ತದೆ.
- **ಏಜೆಂಟ್ ಸ್ಕ್ರ್ಯಾಚ್‌ಪ್ಯಾಡ್‌ಗಳು** ಯಾವುದೇ ಸಂಭಾಷಣೆ ಕಡಿತವನ್ನೂ ಸಹಿಸಬಹುದಾದ ಶಾಶ್ವತ ಬಾಹ್ಯ ಸ್ಮರಣೆ ಒದಗಿಸುತ್ತವೆ.

### ನೀವು ನಿರ್ಮಿಸಿರುವುದು
- ಬಹುಟರ ಸಂಭಾಷಣೆಗಳಾದಲ್ಲಿ ಸತತತೆಯನ್ನು ಕಾಪಾಡುವ **ಸಂದರ್ಭ-ಜಾಗೃತ ಏಜೆಂಟ್**
- ಮುಖ್ಯ ಬಳಕೆದಾರ ವಿವರಗಳನ್ನು ಸಂಕ್ಷಿಪ್ತ ರೂಪದಲ್ಲಿ ದಾಖಲು ಮಾಡಲು **ಸಾರಾಂಶ ಸಾಧನ** (`summarize_preferences`)
- ಸಂಪರ್ಕವನ್ನು ಉಳಿಸುವಿಕೆ ಮತ್ತು ಬದಲಾವಣೆ ನಿರ್ವಹಣೆಯನ್ನು ತೋರಿಸುವ **ಬಹುಟರ ಸಂಭಾಷಣೆ**

### ನೈಜ ಜಾಗತಿಕ ಅನ್ವಯಿಕೆಗಳು
- **ಗ್ರಾಹಕ ಸೇವಾ ಬಾಟ್‌ಗಳು**: ಹದಿನಾರು ಬೆಂಬಲ ಸೆಷನ್‌ಗಳಲ್ಲಿ ಮೆಚ್ಚುಗೆಗಳನ್ನು ನೆನಪಿಡು
- **ವೈಯಕ್ತಿಕ ಸಹಾಯಕರು**: ಪ್ರಸ್ತುತ ಯೋಜನೆಗಳನ್ನು ಮತ್ತೆ ವಿವರಿಸದೆ ಹಿಂಬಾಲಿಸು
- **ಶೈಕ್ಷಣಿಕ ಟ್ಯೂಟರ್ಸ್**: ಹಲವು ಸಂವಾದಗಳಾದಲ್ಲಿ ವಿದ್ಯಾರ್ಥಿಗಳ ಪ್ರಗತಿಯನ್ನು ಕಾಪಾಡಿಕೊಳ್ಳು

### ಮುಂದಿನ ಹೊರಟುಗಳು
- ಕಡತ ಆಧಾರಿತ ಶಾಶ್ವತತೆಯೊಂದಿಗೆ ಸಂಪೂರ್ಣ ಸ್ಕ್ರ್ಯಾಚ್‌ಪ್ಯಾಡ್ ಸಾಧನವನ್ನು ಜಾರಿಗೆ ತರಬೇಕು
- ಸಾರಾಂಶದ ನಂತರ ಸ್ವಯಂಚಾಲಿತ ಇತಿಹಾಸ ಕಡಿತವು ಸೇರಿಸಬೇಕು
- ಸಾಂದರ್ಭಿಕ ಸ್ಮೃತಿ ಸಂಶೋಧನಿಗಾಗಿ ವೆಕ್ಟರ್ ಡೇಟಾಬೇಸ್‌ಗಳ ಜೊತೆ ಸಂಯೋಜಿಸಬೇಕು
- ಸಂಪೂರ್ಣ ಸಂಕಲೆಯಲ್ಲಿ ದಿನಗಳು ಕಳೆದ ಮೇಲೆ ಸಂಭಾಷಣೆಗಳನ್ನು ಪುನರಾರಂಭಿಸಬಹುದಾದ ಏಜೆಂಟ್‌ಗಳನ್ನು ನಿರ್ಮಿಸಬೇಕು
