ಪ್ರೇಕ್ಷಕರು:
- ಇದು ಯಾರಿಗಾಗಿ ಎಂಬುದನ್ನು ವಿವರಿಸಿ.

ಪೂರ್ವಅಗತ್ಯತೆಗಳು:
- ಅಗತ್ಯವಿರುವ ತತ್ತ್ವಗಳು ಅಥವಾ ಸೆಟ್‌ಅಪ್ ಅನ್ನು ಪಟ್ಟಿ ಮಾಡಿ.

ಕಲಿಕೆಯ ಗುರಿಗಳು:
- ಕೊನೆಯಲ್ಲಿ, ಓದುಗನು ... ಮಾಡಲು ಸಾಧ್ಯವಾಗುತ್ತದೆ.


## ರೂಪರೇಖೆ

1. ಸಿದ್ಧತೆ
2. ಕನಿಷ್ಠ ಕಾರ್ಯನಿರ್ವಹಿಸುವ ಉದಾಹರಣೆ
3. ವ್ಯತ್ಯಾಸಗಳು ಮತ್ತು ಅಡಚಣೆಗಳು
4. ಅಭ್ಯಾಸಗಳು


```python
# Setup cell: keep it short and deterministic
from __future__ import annotations

import math
import random

SEED = 21
random.seed(SEED)
SEED
```

## ಹಂತ 1 - ಸಣ್ಣ ಉದಾಹರಣೆಯಿಂದ ಪ್ರಾರಂಭಿಸಿ

ಮುಂದಿನ ಸೆಲ್ ಏನು ಮಾಡುತ್ತದೆ ಎಂಬುದನ್ನು ಸರಳ ಭಾಷೆಯಲ್ಲಿ ವಿವರಿಸಿ.


```python
# Minimal working example
angles = [0, math.pi / 4, math.pi / 2]
sines = [math.sin(a) for a in angles]
list(zip(angles, sines))
```

## ಅಭ್ಯಾಸಗಳು

- ಬೇರೆ ಇನ್‌ಪುಟ್ ಪ್ರಯತ್ನಿಸಿ.
- ಕೋಡ್ ನಡೆಸುವ ಮೊದಲು ಫಲಿತಾಂಶವನ್ನು ಊಹಿಸಿ.
- ಒಂದು ಸಾಮಾನ್ಯ ತಪ್ಪನ್ನು ಮತ್ತು ಅದನ್ನು ಹೇಗೆ ಸರಿಪಡಿಸುವುದು ಎಂಬುದನ್ನು ಗಮನಿಸಿ.


```python
# Exercise answer scaffold
def describe(values: list[float]) -> dict[str, float]:
    return {"min": min(values), "max": max(values)}

describe(sines)
```
