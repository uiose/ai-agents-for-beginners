# 🔍 Microsoft Foundry (.NET) ಜೊತೆಗೆ ಉದ್ಯಮ RAG

## 📋 ಕಲಿಕೆಯ ಉದ್ದೇಶಗಳು

ಈ ನೋಟ್‌ಬುಕ್ Microsoft Foundry ಜೊತೆಗೆ .NET ನಲ್ಲಿ Microsoft Agent Framework ಅನ್ನು ಬಳಸಿಕೊಂಡು ಉದ್ಯಮ-ಗ್ರೇಡ್ Retrieval-Augmented Generation (RAG) ವ್ಯವಸ್ಥೆಗಳನ್ನು ಹೇಗೆ ನಿರ್ಮಿಸುವುದನ್ನು ತೋರಿಸುತ್ತದೆ. ನೀವು ಡಾಕ್ಯುಮೆಂಟ್‌ಗಳಲ್ಲಿ ಹುಡುಕಬಲ್ಲ, ಖಚಿತ ಮತ್ತು ಪ್ರಸಂಗಾನುಕೂಲ ಉತ್ತರಗಳನ್ನು ನೀಡಬಲ್ಲ, ಉತ್ಪಾದನೆಗೆ ಸಿದ್ಧವಾದ ಏಜೆಂಟ್‌ಗಳನ್ನು ಉದ್ಯಮ ಮಟ್ಟದ ಭದ್ರತೆ ಮತ್ತು ವಿಸ್ತರಣಾಶೀಲತೆಯೊಂದಿಗೆ ಹೇಗೆ ರಚಿಸುವುದನ್ನು ಕಲಿಯುತ್ತೀರಿ.

**ನೀವು ನಿರ್ಮಿಸಬಲ್ಲ ಉದ್ಯಮ RAG ಸಾಮರ್ಥ್ಯಗಳು:**
- 📚 **ದಾಖಲೆ ಬುದ್ಧಿವಂತಿಕೆ**: Azure AI ಸೇವೆಗಳ ಮೂಲಕ ಉನ್ನತ ಮಟ್ಟದ ದಾಖಲೆ ಸಂಸ್ಕರಣೆ
- 🔍 **ಅರ್ಥಾತ್ಮಕ ಶೋಧ**: ಉದ್ಯಮ ವೈಶಿಷ್ಟ್ಯಗಳೊಂದಿಗೆ ಉನ್ನತ ಪ್ರದರ್ಶನದ ವೆಕ್ಟರ್ ಶೋಧ
- 🛡️ **ಭದ್ರತೆ ಸಂಯೋಜನೆ**: ಭೂಮಿಕೆಯ ಆಧಾರದ ಪ್ರವೇಶ ಮತ್ತು ಡೇಟಾ ರಕ್ಷಣಾ ಮಾದರಿಗಳು
- 🏢 **ವಿಸ್ತರಣಾಶೀಲ ವಾಸ್ತುಶಿಲ್ಪ**: ಮಾನಿಟರಿಂಗ್‌ೊಂದಿಗೆ ಉತ್ಪಾದನೆಗೆ ಸಿದ್ಧ RAG ವ್ಯವಸ್ಥೆಗಳು

## 🎯 ಉದ್ಯಮ RAG ವಾಸ್ತುಶಿಲ್ಪ

### ಪ್ರಮುಖ ಉದ್ಯಮ ಘಟಕಗಳು
- **Microsoft Foundry**: ಭದ್ರತೆ ಮತ್ತು ಅನುಪಾಲನೆಯೊಂದಿಗೆ ನಿರ್ವಹಿಸಲಾದ ಉದ್ಯಮ AI ವೇದಿಕೆ
- **Persistent Agents**: ಸಂಭಾಷಣೆ ಇತಿಹಾಸ ಮತ್ತು ಪ್ರಸಂಗ ನಿರ್ವಹಣೆಯೊಂದಿಗೆ ಸ್ಥಿತಿಸಹಿತ ಏಜೆಂಟ್‌ಗಳು
- **Vector Store Management**: ಉದ್ಯಮ-ಮಟ್ಟದ ದಾಖಲೆ ಸೂಚ್ಯಂಕೀಕರಣ ಮತ್ತು ಪ್ರಾಪ್ತಿ
- **Identity Integration**: Azure AD ಪ್ರಾಮಾಣೀಕರಣ ಮತ್ತು ಭೂಮಿಕೆಯ ಆಧಾರದ ಪ್ರವೇಶ ನಿಯಂತ್ರಣ

### .NET ಉದ್ಯಮ ಪ್ರಯೋಜನಗಳು
- **Type Safety**: RAG ಕಾರ್ಯಾಚರಣೆಗಳು ಮತ್ತು ಡೇಟಾ ರಚನೆಗಳಿಗೆ ಸಂಕಲನ-ಸಮಯ ಮಾನ್ಯತೆ
- **Async Performance**: ತಡೆಗಟ್ಟದೆ ಕಾರ್ಯನಿರ್ವಹಿಸುವ ದಾಖಲೆ ಪ್ರಕ್ರಿಯೆ ಮತ್ತು ಶೋಧ ಕಾರ್ಯಗಳು
- **Memory Management**: ದೊಡ್ಡ ದಾಕ್ಯುಮೆಂಟ್ ಸಂಗ್ರಹಗಳಿಗಾಗಿ ಸಂಪನ್ಮೂಲಗಳ ಪರಿಣಾಮಕಾರಿ ಉಪಯೋಗ
- **Integration Patterns**: ಡಿಪೆಂಡೆನ್ಸಿ ಇಂಜೆಕ್ಷನ್‌ನೊಂದಿಗೆ ಸ್ಥಳೀಯ Azure ಸೇವೆ ಏಕೀಕರಣ ಮಾದರಿಗಳು

## 🏗️ ತಾಂತ್ರಿಕ ವಾಸ್ತುಶಿಲ್ಪ

### ಉದ್ಯಮ RAG ಪೈಪ್‍ಲೈನ್
```csharp
Document Upload → Security Validation → Vector Processing → Index Creation
                      ↓                    ↓                  ↓
User Query → Authentication → Semantic Search → Context Ranking → AI Response
```

### ಪ್ರಮುಖ .NET ಘಟಕಗಳು
- **Azure.AI.Agents.Persistent**: ಸ್ಥಿತಿ ಸಂರಕ್ಷಣೆಯೊಂದಿಗೆ ಉದ್ಯಮ ಏಜೆಂಟ್ ನಿರ್ವಹಣೆ
- **Azure.Identity**: ಸುರಕ್ಷಿತ Azure ಸೇವೆ ಪ್ರವೇಶಕ್ಕಾಗಿ ಏಕೀಕೃತ ಪ್ರಾಮಾಣೀಕರಣ
- **Microsoft.Agents.AI.AzureAI**: Azure-ಅನುಕೂಲಿತ ಏಜೆಂಟ್ ಫ್ರೇಮ್ವರ್ಕ್ ಅನುಷ್ಠಾನ
- **System.Linq.Async**: ಉನ್ನತ-ಕಾರ್ಯಕ್ಷಮತೆಯ ಅಸಿಂಕ್ರೋನಸ್ LINQ ಕಾರ್ಯಾಚರಣೆಗಳು

## 🔧 ಉದ್ಯಮ ವೈಶಿಷ್ಟ್ಯಗಳು ಮತ್ತು ಪ್ರಯೋಜನಗಳು

### ಭದ್ರತೆ ಮತ್ತು ಅನುಪಾಲನೆ
- **Azure AD Integration**: ಉದ್ಯಮ ಗುರುತು ನಿರ್ವಹಣೆ ಮತ್ತು ಪ್ರಾಮಾಣೀಕರಣ
- **Role-Based Access**: ದಾಖಲೆ ಪ್ರವೇಶ ಮತ್ತು ಕಾರ್ಯಚಟುವಟಿಕೆಗಳಿಗೆ ಸೂಕ್ಷ್ಮ ಹಕ್ಕು ನಿಯಂತ್ರಣ
- **Data Protection**: ಸಂವೇದನಾಶೀಲ ದಾಖಲೆಗಳಿಗೆ ವಿಶ್ರಾಂತದಲ್ಲಿಯೂ ಮತ್ತು ಸಾರಣಿಯಲ್ಲಿ ಎನ್‌ಕ್ರಿಪ್ಶನ್
- **Audit Logging**: ಅನುಪಾಲನೆ ಅಗತ್ಯಗಳಿಗೆ ಸಮಗ್ರ ಚಟುವಟಿಕೆ ಟ್ರ್ಯಾಕಿಂಗ್

### ಕಾರ್ಯಕ್ಷಮತೆ ಮತ್ತು ವಿಸ್ತರಣಾಶೀಲತೆ
- **Connection Pooling**: Azure ಸೇವೆ ಸಂಪರ್ಕಗಳ ಪರಿಣಾಮಕಾರಿ ನಿರ್ವಹಣೆ
- **Async Processing**: ಹೆಚ್ಚಿನ ಪ್ರಮಾಣದ ಪರಿಸ್ಥಿತಿಗಳಿಗೆ ತಡೆಗಟ್ಟದೆ ಕಾರ್ಯನಿರ್ವಹಣೆ
- **Caching Strategies**: ಹೆಚ್ಚು ಪ್ರವೇಶಿಸಲ್ಪಡುವ ದಾಖಲೆಗಳಿಗೆ ಬುದ್ಧಿವಂತ క్యಾಶಿಂಗ್
- **Load Balancing**: ದೊಡ್ಡ-ಗಾತ್ರದ ನಿಯೋಜನೆಗಳಿಗಾಗಿ ವಿತರಿತ ಪ್ರಕ್ರಿಯೆ

### ನಿರ್ವಹಣೆ ಮತ್ತು ಮಾನಿಟರಿಂಗ್
- **Health Checks**: RAG ವ್ಯವಸ್ಥೆ ಘಟಕಗಳಿಗಾಗಿ ಒಳಗೊಂಡ ಮಾನಿಟರಿಂಗ್
- **Performance Metrics**: ಶೋಧ ಗುಣಮಟ್ಟ ಮತ್ತು ಪ್ರತಿಕ್ರಿಯಾ ಸಮಯದ ಬಗ್ಗೆ ವಿವರವಾದ ವಿಶ್ಲೇಷಣೆ
- **Error Handling**: ಮರುಪ್ರಯತ್ನ ನೀತಿಗಳೊಂದಿಗೆ ಸಮಗ್ರ ಎಕ್ಸೆಪ್‌ಷನ್ ನಿರ್ವಹಣೆ
- **Configuration Management**: ವಾತಾವರಣ-ನಿರ್ದಿಷ್ಟ ಸೆಟ್ಟಿಂಗ್‌ಗಳು ಮತ್ತು ಮಾನ್ಯತೆ

## 📊 ಉದ್ಯಮ RAG ಮಾದರಿಗಳು

### ದಾಖಲೆ ನಿರ್ವಹಣೆ ಮಾದರಿಗಳು
- **Bulk Upload**: ದೊಡ್ಡ ದಾಖಲೆ ಸಂಗ್ರಹಗಳ ಪರಿಣಾಮಕಾರಿ ಪ್ರಕ್ರಿಯೆ
- **Incremental Updates**: ರಿಯಲ್‌ಟೈಮ್ ದಾಖಲೆ ಸೇರಿಸುವಿಕೆ ಮತ್ತು ಬದಲಾವಣೆ
- **Version Control**: ದಾಖಲೆ ಆವೃತ್ತಿ ನಿಯಂತ್ರಣ ಮತ್ತು ಬದಲಾವಣೆಯ ಟ್ರ್ಯಾಕಿಂಗ್
- **Metadata Management**: ಸಮೃದ್ಧ ದಾಖಲೆ ಗುಣಲಕ್ಷಣಗಳು ಮತ್ತು ವರ್ಗೀಕರಣ

### ಶೋಧ ಮತ್ತು ಪ್ರಾಪ್ತಿ ಮಾದರಿಗಳು
- **Hybrid Search**: ಉತ್ತಮ ಫಲಿತಾಂಶಗಳಿಗೆ ಅರ್ಥಾತ್ಮಕ ಮತ್ತು ಕೀವರ್ಡ್ ಶೋಧವನ್ನು ಸಂಯೋಜಿಸುವುದು
- **Faceted Search**: ಬಹು-ಆಯಾಮ ಫಿಲ್ಟರಿಂಗ್ ಮತ್ತು ವರ್ಗೀಕರಣ
- **Relevance Tuning**: ಕ್ಷೇತ್ರ-ನಿರ್ದಿಷ್ಟ ಅಗತ್ಯಗಳಿಗೆ ಕಸ್ಟಮ್ اسڪೋರಿಂಗ್ ಅಲ್ಗೋರಿದಮ್‌ಗಳು
- **Result Ranking**: ವಹಿವಾಟು ತರ್ಕ ಏಕೀಕರಣದೊಂದಿಗೆ ಸುಧಾರಿತ ರ್ಯಾಂಕಿಂಗ್

### ಭದ್ರತಾ ಮಾದರಿಗಳು
- **Document-Level Security**: ಪ್ರತಿ ದಾಖಲೆಯಾಗಿ ಸೂಕ್ಷ್ಮ ಪ್ರವೇಶ ನಿಯಂತ್ರಣ
- **Data Classification**: ಸ್ವಯಂಚಾಲಿತ ಸಂವೇದನಾಶೀಲತೆ ಲೇಬಲಿಂಗ್ ಮತ್ತು ರಕ್ಷಣೆ
- **Audit Trails**: ಎಲ್ಲಾ RAG ಕಾರ್ಯಾಚರಣೆಗಳ ಸಮಗ್ರ ಲಾಗಿಂಗ್
- **Privacy Protection**: PII ಪತ್ತೆ ಮತ್ತು ರಿಡ್ಯಾಕ್ಷನ್ ಸಾಮರ್ಥ್ಯಗಳು

## 🔒 ಉದ್ಯಮ ಭದ್ರತಾ ವೈಶಿಷ್ಟ್ಯಗಳು

### ಪ್ರತಿಮಾಣೀಕರಣ ಮತ್ತು ಅನುಮತಿ
```csharp
// Azure AD integrated authentication
var credential = new AzureCliCredential();
var agentsClient = new PersistentAgentsClient(endpoint, credential);

// Role-based access validation
if (!await ValidateUserPermissions(user, documentId))
{
    throw new UnauthorizedAccessException("Insufficient permissions");

```

### ಡೇಟಾ ರಕ್ಷಣೆ
- **Encryption**: ದಾಖಲೆಗಳು ಮತ್ತು ಶೋಧ ಸೂಚ್ಯಂಕಗಳಿಗಾಗಿ ಎಂಡ್-ಟು-ಎಂಡ್ ಎನ್‌ಕ್ರಿಪ್ಷನ್
- **Access Controls**: ಬಳಕೆದಾರ ಮತ್ತು ಗುಂಪು ಹಕ್ಕುಗಳಿಗಾಗಿ Azure AD ಜೊತೆಗೆ ಏಕೀಕರಣ
- **Data Residency**: ಅನುಪಾಲನೆಗಾಗಿ ಭೌಗೋಳಿಕ ಡೇಟಾ ಸ್ಥಳ ನಿಯಂತ್ರಣ
- **Backup & Recovery**: ಸ್ವಯಂಚಾಲಿತ ಬ್ಯಾಕಪ್ ಮತ್ತು ವಿಪತ್ತು ಪುನರ್‌ರಕ್ಷಣೆ ಸಾಮರ್ಥ್ಯಗಳು

## 📈 ಕಾರ್ಯಕ್ಷಮತೆ ಸುಧಾರಣೆ

### ಅಸಿಂಕ್ ಪ್ರೊಸೆಸಿಂಗ್ ಮಾದರಿಗಳು
```csharp
// Efficient async document processing
await foreach (var document in documentStream.AsAsyncEnumerable())
{
    await ProcessDocumentAsync(document, cancellationToken);
}
```

### ಮೆಮರಿ ನಿರ್ವಹಣೆ
- **Streaming Processing**: ಮೆಮರಿ ಸಮಸ್ಯೆಗಳಿಲ್ಲದೆ ದೊಡ್ಡ ದಾಖಲೆಗಳನ್ನು ಪ್ರಕ್ರಿಯೆಗೊಳಿಸುವುದು
- **Resource Pooling**: ಖರ್ಚುಬರಿತದ ಸಂಪನ್ಮೂಲಗಳ ಪರಿಣಾಮಕಾರಿ ಮರುಉಪಯೋಗ
- **Garbage Collection**: ಮೆಮರಿ ಹಂಚಿಕೆಗೆ ಪರಿಷ್ಕೃತ ಮಾದರಿಗಳು
- **Connection Management**: ಸರಿಯಾದ Azure ಸೇವೆ ಸಂಪರ್ಕ ಜೀವನಚಕ್ರ ನಿರ್ವಹಣೆ

### ಕ್ಯಾಶಿಂಗ್ ತಂತ್ರಗಳು
- **Query Caching**: ಹೆಚ್ಚು ನಡೆಸಲ್ಪಡುವ ಶೋಧಗಳನ್ನು ಕ್ಯಾಶ್ ಮಾಡುವುದು
- **Document Caching**: ಹಾಟ್ ದಾಖಲೆಗಳಿಗಾಗಿ ಮೆಮರಿಯಲ್ಲಿ ಕ್ಯಾಶಿಂಗ್
- **Index Caching**: ವೆಕ್ಟರ್ ಸೂಚ್ಯಂಕ ಕ್ಯಾಶಿಂಗ್ ಪರಿಷ್ಕೃತ
- **Result Caching**: ಜನರೇಟ್ ಆಗಿದ ಉತ್ತರಗಳ ಬುದ್ಧಿವಂತ ಕ್ಯಾಶಿಂಗ್

````polyglot-notebook
using System;
using System.Linq;
using Azure.AI.Agents.Persistent;
using Azure.Identity;
using Microsoft.Agents.AI;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
var azure_foundry_endpoint = Environment.GetEnvironmentVariable("AZURE_AI_PROJECT_ENDPOINT") ?? throw new InvalidOperationException("AZURE_AI_PROJECT_ENDPOINT is not set.");
var azure_foundry_model_id = Environment.GetEnvironmentVariable("AZURE_AI_MODEL_DEPLOYMENT_NAME") ?? "gpt-4.1-mini";
````

````polyglot-notebook
string pdfPath = "./document.md";
````

````polyglot-notebook
using System.IO;

async Task<Stream> OpenImageStreamAsync(string path)
{
	return await Task.Run(() => File.OpenRead(path));
}

var pdfStream = await OpenImageStreamAsync(pdfPath);
````

````polyglot-notebook
var persistentAgentsClient = new PersistentAgentsClient(azure_foundry_endpoint, new AzureCliCredential());
````

````polyglot-notebook
PersistentAgentFileInfo fileInfo = await persistentAgentsClient.Files.UploadFileAsync(pdfStream, PersistentAgentFilePurpose.Agents, "demo.md");
````

````polyglot-notebook
PersistentAgentsVectorStore fileStore =
            await persistentAgentsClient.VectorStores.CreateVectorStoreAsync(
                [fileInfo.Id],
                metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
PersistentAgent agentModel = await persistentAgentsClient.Administration.CreateAgentAsync(
            azure_foundry_model_id,
            name: "DotNetRAGAgent",
            tools: [new FileSearchToolDefinition()],
            instructions: """
                You are an AI assistant designed to answer user questions using only the information retrieved from the provided document(s).

                - If a user's question cannot be answered using the retrieved context, **you must clearly respond**: 
                "I'm sorry, but the uploaded document does not contain the necessary information to answer that question."
                - Do not answer from general knowledge or reasoning. Do not make assumptions or generate hypothetical explanations.
                - Do not provide definitions, tutorials, or commentary that is not explicitly grounded in the content of the uploaded file(s).
                - If a user asks a question like "What is a Neural Network?", and this is not discussed in the uploaded document, respond as instructed above.
                - For questions that do have relevant content in the document (e.g., Contoso's travel insurance coverage), respond accurately, and cite the document explicitly.

                You must behave as if you have no external knowledge beyond what is retrieved from the uploaded document.
                """,
            toolResources: new()
            {
                FileSearch = new()
                {
                    VectorStoreIds = { fileStore.Id },
                }
            },
            metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
AIAgent agent = await persistentAgentsClient.GetAIAgentAsync(agentModel.Id);
````

````polyglot-notebook
AgentThread thread = agent.GetNewThread();
````

````polyglot-notebook
Console.WriteLine(await agent.RunAsync("Can you explain Contoso's travel insurance coverage?", thread));
````
