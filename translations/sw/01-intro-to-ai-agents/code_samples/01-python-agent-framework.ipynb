```python
import logging
logging.getLogger("agent_framework.azure").setLevel(logging.ERROR)

import os
import asyncio
from typing import Annotated

from agent_framework import tool
from agent_framework.azure import AzureAIProjectAgentProvider
from azure.identity import AzureCliCredential

provider = AzureAIProjectAgentProvider(credential=AzureCliCredential())
```

## Kuunda Wakala Wako wa Kwanza

Wakala anahitaji vitu viwili:

- **Maelekezo** yanayomwambia *ni nani* na *jinsi ya tabia* (amri ya mfumo).
- **Zana** — kazi za Python zilizo na alama `@tool` ambazo wakala anaweza kuitumia kupata taarifa au kufanya vitendo.

Hapo chini tunaelezea zana rahisi inayorudisha orodha ya maeneo maarufu ya likizo. Wakala atatumia zana hii wakati mtumiaji atauliza mapendekezo ya kusafiri.


```python
@tool(approval_mode="never_require")
def get_destinations() -> list[str]:
    """Get a list of popular vacation destinations."""
    return [
        "Barcelona",
        "Paris",
        "Berlin",
        "Tokyo",
        "Sydney",
        "New York City",
        "Cairo",
        "Cape Town",
        "Rio de Janeiro",
        "Bali",
    ]
```

```python
agent = await provider.create_agent(
    tools=[get_destinations],
    name="TravelAgent",
    instructions=(
        "You are a helpful travel agent. Help users find their perfect vacation "
        "destination based on their preferences. Use the get_destinations tool "
        "to see available destinations."
    ),
)

response = await agent.run(
    "I'm looking for a warm beach destination. What do you recommend?"
)
print(response)
```

## Majibu yanayotiririka

Kwa uzoefu wa mwingiliano zaidi unaweza **kutitiririsha** jibu la wakala. Badala ya kusubiri jibu kamili, wakala hutawanya vipande vya maandishi vinavyotengenezwa. Hii ni muhimu hasa katika mwingiliano wa mazungumzo ambapo unataka kuonyesha matokeo kwa wakati halisi.


```python
async for chunk in agent.run(
    "Tell me about Tokyo as a travel destination", stream=True
):
    print(chunk, end="", flush=True)
```
